```python
import numpy as np
import sympy as sp
import pandas as pd
import math
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split


data=pd.read_csv('C:\\Users\\Utsav\\Desktop\\pulsar_prediction\\pulsar_stars.csv', sep=',',header=0)


pulsarData=data.values
pulsarData=np.array(pulsarData)
```

```python
#split the data

train, test = train_test_split(data, test_size=0.2)

train=np.array(train)
test=np.array(test)

# k fold cross validation on training set  (k=3)

#divide the training set into 3 sets (set1, set2, set3) using test_train_split twice

set1, set2_3 = train_test_split(train, test_size=0.66666666666)

set2, set3 = train_test_split(set2_3, test_size=0.5)

print(len(set1))
print(len(set2))
print(len(set3))

print(len(test))
print(len(train))
```

```python
#copy labels of test & train data elsewhere
y1=np.zeros(len(set1))
y2=np.zeros(len(set2))
y3=np.zeros(len(set3))
y_test=np.zeros(len(test))
for i in range(len(test)):
    y_test[i]=test[i][8]
    test[i][8]=-1

for i in range(len(set1)):
    y1[i]=set1[i][8]
    set1[i][8]=-1
    
for i in range(len(set2)):
    y2[i]=set2[i][8]
    set2[i][8]=-1
    
for i in range(len(set3)):
    y3[i]=set3[i][8]
    set3[i][8]=-1
```

```python
#import knn classifier from sklearn

from sklearn.neighbors import KNeighborsClassifier
```

```python
"""
Fit the model using set1 and set3. Test this model on set2. Call this model2
"""

model2 = KNeighborsClassifier(n_neighbors=5)
#first fit on set1
model2.fit(set1, y1)
#then fit on set3
model2.fit(set3, y3)

#Now test (validate) on set2
y_pred2 = model2.predict(set2)

#calculate score 
score=0
for i in range(len(set2)):
    if y_pred2[i]==y2[i]:
        score=score+1
score=score*100/len(set2)

print("Accuracy = " + "%.2f" % score + "%")

```

```python
"""
Fit the model using set1 and set2. Test this model on set3. Call this model3
"""

model3 = KNeighborsClassifier(n_neighbors=5)
#first fit on set1
model3.fit(set1, y1)
#then fit on set2
model3.fit(set2, y2)

#Now test (validate) on set1
y_pred3 = model3.predict(set3)

#calculate score 
score=0
for i in range(len(set3)):
    if y_pred3[i]==y3[i]:
        score=score+1
score=score*100/len(set3)

print("Accuracy = " + "%.2f" % score + "%")
```

```python
"""
Now....
On the test set, predict using all three models and make final prediction by majority vote

"""

y_pred_m1 = model1.predict(test)
y_pred_m2 = model2.predict(test)
y_pred_m3 = model3.predict(test)
y_pred = np.zeros(len(test))

#taking majority vote
for i in range(len(test)):
    pred=y_pred_m1[i]+y_pred_m2[i]+y_pred_m3[i]
        
    if(pred == 0):
        y_pred[i]=0
    if(pred == 1):
        y_pred[i]=0
    if(pred == 2):
        y_pred[i]=1
    if(pred == 3):
        y_pred[i]=1
    
```

```python
#calculate score 
score=0
for i in range(len(test)):
    if y_pred[i]==y_test[i]:
        score=score+1
score=score*100/len(test)

print("Accuracy = " + "%.2f" % score + "%")

#calculate no of flase negatives
score=0
pos=0
for i in range(len(test)):
    if y_pred[i]==0 and y_test[i]==1:
        score=score+1
    if y_test[i]==1:
        pos=pos+1
        
#score=score*100/len(test)
percent = score/len(test)


print("No of false negatives = %d" % score)
print("No of positives = %d" % pos)
```

To decrease the number of false negatives, Let's modify kNN little bit.
Now, even if one neighbor is a Pulsar, we wil classify that datapoint as a pulsar, instead of taking the majority vote.
This wil increase the number of false positives, but thats fine for our application

```python
%%time
#without feature scaling
def ed(arr1, arr2):
    ed = 0
    for i in range(8):
        ed = ed + (arr1[i]-arr2[i])**2
    
    return math.sqrt(ed)



y_pred1=np.zeros(len(test))
y_pred2=np.zeros(len(test))
y_pred3=np.zeros(len(test))
y_pred=np.zeros(len(test))

#Train and predict on set1
for i in range(len(test)):
    dist = np.zeros(len(set1))
    label = np.zeros(len(set1))
    for j in range(len(set1)):
        dist[j]=ed(set1[j], test[i])
        label[j]=y1[j]
        
    l1 = np.argmin(dist)
    dist[l1] = 999999999999
    l1 = label[l1]
    l2 = np.argmin(dist)
    dist[l2] = 999999999999
    l2 = label[l2]
    l3 = np.argmin(dist)
    dist[l3] = 999999999999
    l3 = label[l3]
    
    knn = l1+l2+l3
    if(knn == 0):
        y_pred1[i]=0
    if(knn == 1):
        y_pred1[i]=1
    if(knn == 2):
        y_pred1[i]=1
    if(knn == 3):
        y_pred1[i]=1
        

#Train and predict on set2
for i in range(len(test)):
    dist = np.zeros(len(set2))
    label = np.zeros(len(set2))
    for j in range(len(set2)):
        dist[j]=ed(set2[j], test[i])
        label[j]=y2[j]
        
    l1 = np.argmin(dist)
    dist[l1] = 999999999999
    l1 = label[l1]
    l2 = np.argmin(dist)
    dist[l2] = 999999999999
    l2 = label[l2]
    l3 = np.argmin(dist)
    dist[l3] = 999999999999
    l3 = label[l3]
    
    knn = l1+l2+l3
    if(knn == 0):
        y_pred2[i]=0
    if(knn == 1):
        y_pred2[i]=1
    if(knn == 2):
        y_pred2[i]=1
    if(knn == 3):
        y_pred2[i]=1
        

#Train and predict on set3
for i in range(len(test)):
    dist = np.zeros(len(set3))
    label = np.zeros(len(set3))
    for j in range(len(set3)):
        dist[j]=ed(set3[j], test[i])
        label[j]=y2[j]
        
    l1 = np.argmin(dist)
    dist[l1] = 999999999999
    l1 = label[l1]
    l2 = np.argmin(dist)
    dist[l2] = 999999999999
    l2 = label[l2]
    l3 = np.argmin(dist)
    dist[l3] = 999999999999
    l3 = label[l3]
    
    knn = l1+l2+l3
    if(knn == 0):
        y_pred3[i]=0
    if(knn == 1):
        y_pred3[i]=1
    if(knn == 2):
        y_pred3[i]=1
    if(knn == 3):
        y_pred3[i]=1
        
#taking majority vote
for i in range(len(test)):
    pred=y_pred1[i]+y_pred2[i]+y_pred3[i]
        
    if(pred == 0):
        y_pred[i]=0
    if(pred == 1):
        y_pred[i]=0
    if(pred == 2):
        y_pred[i]=1
    if(pred == 3):
        y_pred[i]=1
            


score=0
for l in range(len(test)):
    if y_pred[l]==y_test[l]:
        score=score+1
score=score*100/len(test)
print("Accuracy = " + "%.2f" % score + "%")

#calculate no of flase negatives
score=0

for i in range(len(test)):
    if y_pred[i]==0 and y_test[i]==1:
        score=score+1

        
#score=score*100/len(test)
percent = score/len(test)


print("No of false negatives = %d" % score)
print("No of positives in test data = %d" % sum(y_test))
print("No of positives predicted : %d" % sum(y_pred))
```
