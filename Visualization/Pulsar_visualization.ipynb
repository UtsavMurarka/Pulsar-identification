```python
import numpy as np
import sympy as sp
import pandas as pd
import math
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split


data=pd.read_csv('C:\\Users\\Utsav\\Desktop\\pulsar_prediction\\pulsar_stars.csv', sep=',',header=0)


pulsarData=data.values
pulsarData=np.array(pulsarData)
```

```python
#split the data

train, test = train_test_split(data, test_size=0.2)
pulsar=np.array(train)
test=np.array(test)
```

```python
train=np.array(train)
NegativeCount=0
PositiveCount=0
for i in range(14318):
    if pulsar[i][8]==0 :
        NegativeCount=NegativeCount+1
    if pulsar[i][8]==1:
        PositiveCount=PositiveCount+1
print(NegativeCount)
print(PositiveCount)
print(PositiveCount+NegativeCount)
```

```python
positives=np.zeros((PositiveCount,9))
negatives=np.zeros((NegativeCount,9))
    
j=0
k=0
    
for i in range(14318):

    if(pulsar[i,8] == 1):
        positives[j,:] = pulsar[i,:]
        j=j+1
    if(pulsar[i,8] == 0):
        negatives[k,:] = pulsar[i,:]
        k=k+1
```

# Mean of IPP

```python
#Mean IPP
bins=20
clr=['r', 'b']
lbl=['1', '0']
x=[positives[:,0], negatives[:,0]]
plt.hist(x, bins, density=True, color=clr, label=lbl)
plt.legend(loc='upper right')
plt.title('Density')
plt.show()
#
bins = np.linspace(-1, 250, 50)
plt.hist(positives[:,0], bins, alpha=0.5, label='1', color='r', density=True)
plt.hist(negatives[:,0], bins, alpha=0.5, label='0', color='b', density=True)
plt.legend(loc='upper right')
plt.title('Density overlap')
plt.show()
#
bins = np.linspace(-1, 250, 50)
plt.hist(positives[:,0], bins, alpha=0.5, label='1', color='r')
plt.hist(negatives[:,0], bins, alpha=0.5, label='0', color='b')
plt.legend(loc='upper right')
plt.title('without Density overlap')
plt.show()
#
bins=100
plt.hist(positives[:,0], bins)
plt.title('only postives (Not density)')
plt.show()
#
bins=100
plt.hist(negatives[:,0],bins)
plt.title('only negatives (Not density)')
plt.show()
#
print("Mean of Mean of IPP for positive data = ", np.mean(positives[:,0]))
print("Mean of Mean of IPP for negative data = ", np.mean(negatives[:,0])) 
```

Inference:<br>
Means are significantly away
<br>
Overlap is not much
